```python
import nltk
nltk.download("vader_lexicon")
```

```python
import numpy as np
import pandas as pd
```

```python
data=pd.read_csv("twitter_training.csv")
```

```python
data.head()
```

```python
data.isnull().sum()
```

```python
data.dropna(inplace=True)
```

```python
blanks=[]
for i,label,Bd,p,rv in data.itertuples():
  if type(rv)==str:
    if rv.isspace():
      blanks.append(i)
blanks
```

```python
data.drop(blanks,inplace=True)
```

```python
data["Positive"].value_counts()
```

```python
from nltk.sentiment.vader import SentimentIntensityAnalyzer
sid=SentimentIntensityAnalyzer()
```

```python
data["scores"]=data["review"].apply(lambda review:sid.polarity_scores(review))
```

```python
data["compound_scores"]=data["scores"].apply(lambda score__dict:score__dict["compound"])
```

```python
data["label"]=data["compound_scores"].apply(lambda c:"Positive" if c>0 else ("Negative" if c < 0 else "Neutral"))
```

```python
data.head()
```

```python
from sklearn.metrics import confusion_matrix,accuracy_score,classification_report
```

```python
cm=confusion_matrix(data["Positive"],data["label"])
print(cm)
```

```python
print(classification_report(data["Positive"],data["label"]))
```

```python
accuracy_score(data["Positive"],data["label"])
```
